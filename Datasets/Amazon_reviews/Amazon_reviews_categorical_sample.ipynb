```python
import pandas as pd
import gzip
```

```python
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)
```

```python
def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDF('reviews_Grocery_and_Gourmet_Food_5.json.gz')
```

```python
df.head()
```

```python
#Building a random sample of tweets
df_sample = df.sample(n=10000, random_state=42)
df_sample
```

```python
df_sample["overall"].value_counts()
```

```python
#Classifying stars - replacing ratings of 1-2 stars as 0 (negative), 3-4 stars as 1 neutral - leaving 5 stars as 5 (positive)
df_sample.loc[(df_sample["overall"] < 3)] = 0
df_sample.loc[(df_sample["overall"] == 3 ) | (df_sample["overall"] == 4), "overall"] = 1

df_sample
```

```python
df_sample["overall"].value_counts()
```

```python
df_final = df_sample[["reviewText", "overall"]]
df_final
```

```python
#Saving sample file to csv
df_final.to_csv('final_sample.csv')
```
